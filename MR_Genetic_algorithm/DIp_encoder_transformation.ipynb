```python
import os
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
```

```python
dir_name = os.path.split(os.getcwd())[0]
path_name = f"{dir_name}\Datasets\R_prepr_2024_08_15.csv"
print(path_name)
```

```python
import numpy as np
import itertools
correl_features = {"Type of reactor": ["electrolyte", "conc, M", "pH"], 
                   'width min (nm)': ['width aver (nm)', 'width max (nm)', 'length min (nm)', 'length aver (nm)',  'length max (nm)'],
                   'Cu, %':['Cu+, %',  'Cu(2+), %']}
a = list(itertools.chain(*correl_features.values()))
a
```

```python
product = "C2H4"
df = pd.read_csv(path_name)
Xx = df.drop(['FE, %', 'Unnamed: 0', 'DOI', 'Cu2O-110, %', 'Cu2O-111, %', 'Cu2O-200, %', 'Cu2O-220, %', 'Cu-111, %', 'Cu-200, %', 'Cu-220, %', 'XRD noises', 'Current density, mA/cm2'], axis=1)
Xx = Xx.loc[Xx["Product"] == product]
Xx["Support"] = Xx["Support"].map("{}-s".format)
data = Xx.copy()
```

```python
categorical_cols = data.select_dtypes(include=['object']).columns
categorical_cols = list(categorical_cols)
categorical_cols
```

```python
encoder = OneHotEncoder()

categorical_cols = data.select_dtypes(include=['object']).columns
buffer = data[categorical_cols].copy()
encoder.fit(buffer)
encoded_df = pd.DataFrame(encoder.transform(buffer).toarray(), columns=list(encoder.get_feature_names_out()))
data = pd.concat([data, encoded_df], axis=1)
data = data.drop(columns = categorical_cols)
data.head(10)
```

```python
scaler = StandardScaler()
normalized_df = pd.DataFrame(scaler.fit_transform(data), columns=data.columns)
```

```python
categorical_cols = data.select_dtypes(include=['object']).columns
for col in categorical_cols:
    buffer = pd.get_dummies(data[col])
    data = pd.concat([data, buffer], axis=1)
    data = data.drop(columns = col)
scaler = StandardScaler()
normalized_df = pd.DataFrame(scaler.fit_transform(data), columns=data.columns)
```
